```python
import albumentations as A
import random

transform = A.Compose([
        A.HorizontalFlip(p=0.5),
        A.OneOf([
            A.GaussNoise(),
            A.GaussNoise(),
        ], p=0.2),
        A.OneOf([
            A.MotionBlur(p=.2),
            A.MedianBlur(blur_limit=3, p=0.1),
            A.Blur(blur_limit=3, p=0.1),
        ], p=0.1),
        A.ShiftScaleRotate(shift_limit=0.0625, scale_limit=0.2, rotate_limit=20, p=0.4),
        A.OneOf([
            A.OpticalDistortion(p=0.2),
            A.GridDistortion(p=.1),
        ], p=0.2),
        A.OneOf([
            A.CLAHE(clip_limit=2),
            A.Sharpen(),
            A.Emboss(),
            A.RandomBrightnessContrast(),            
        ], p=0.3),
        A.HueSaturationValue(p=0.2),
    ])
```

```python
import cv2
import os
from os import listdir
import glob
face_detector = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')


for img1 in glob.glob("img/*.jpg"):#Accessing the images in img folder one by one
    n= cv2.imread(img1)

    img = n
    #Gray Scaling of image
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_detector.detectMultiScale(gray, 1.3, 2)
    
    xy=img1.split("\\")[1]
    xx=xy.split(".")[0] 
    
    for (x,y,w,h) in faces:

        cv2.rectangle(img, (x,y), (x+w,y+h), (0,255,0), 3)     

        # Save the captured image into the datasets folder
        cv2.imwrite("dataset/User." + xx + '.' + str(0) + ".jpg", gray[y:y+h,x:x+w])      
    
```

```python
from tqdm import tqdm 

image_dir = "dataset"
all_classes = sorted(os.listdir(image_dir)) 
list1=[]


f = open("name.txt", 'a+')

#Getting all id's present 
for i in range(len(all_classes)):
    list1.append(all_classes[i].split(".")[1])
list1.sort(key = int)

#Getting the last id 
x=list1[len(list1)-1]
y=int(x)

#Adding ids' in name.txt file 
for i in range(y+1):
    f.write(str(i) + '\n')

f.close()
display_images = []

# Create augmented version of each image
for class_name in all_classes:
    
    count = 1  
    
    # read image from disk
    image1 = cv2.imread(os.path.join(image_dir, class_name))
    
    # save 10 augmentation for this image
    for i in tqdm(range(10)):
        augmented_image = transform(image=image1)['image']
        cv2.imwrite("dataset/User." + class_name.split(".")[1]+ '.' + str(count) + ".jpg", augmented_image)
        count=count+1
```
